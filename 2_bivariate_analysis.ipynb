```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

df = pd.read_csv('datasets/merged_dataset.csv')
```

```python
num_cols = df.select_dtypes(include=['number']).columns.tolist()
num_cols.remove('_id')
num_cols.remove('asset_allocation_id')

cat_cols = df.select_dtypes(include=['object']).columns.tolist()
cat_cols.remove('created')
```

```python
sns.pairplot(df[num_cols], corner=True)
plt.suptitle('Scatter plots between numerical features', y=1.02)
plt.tight_layout()
plt.show()
```

```python
for num in num_cols:
    for cat in cat_cols:
        plt.figure(figsize=(10, 4))
        sns.violinplot(x=cat, y=num, data=df, inner='box', palette='Set2')  # sin box interno
        sns.stripplot(x=cat, y=num, data=df, color='k', alpha=0.3, jitter=True, size=2)  # puntos individuales
        plt.title(f'{num} vs {cat}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```

```python
# BONUS: TESTING CONVERSION RATES HYPOTHESIS
# This section tests the hypothesis that conversing asset values to GBP will yield a less disperse distribution of values.

conversion_rates = {
    'GBP': 1.00,
    'USD': 0.79,
    'EUR': 0.85,
    'JPY': 0.0058,
    'AUD': 0.52
}

df['asset_value_gbp'] = df.apply(lambda row: row['asset_value'] * conversion_rates.get(row['asset_currency'], 1.0), axis=1)

plt.figure(figsize=(10, 5))
sns.histplot(df['asset_value_gbp'], bins=50, kde=True)
plt.title('Distribution of asset_value converted to GBP')
plt.xlabel('Asset Value (GBP)')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 4))
sns.violinplot(x='asset_currency', y='asset_value_gbp', data=df, inner='box', palette='Set2')  
sns.stripplot(x='asset_currency',y='asset_value_gbp', data=df, color='k', alpha=0.3, jitter=True, size=2)
plt.title(f'asset_value_gbp vs asset_currency')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
num_cols.append('asset_value_gbp')

sns.pairplot(df[num_cols], corner=True)
plt.suptitle('Scatter plots between numerical features (asset_value_gbp added)', y=1.02)
plt.tight_layout()
plt.show()
```

```python
contingency_table = pd.crosstab(df['asset_allocation'], df['asset_currency'])
print("Contingency Table of Asset Allocation by Currency:")
print(contingency_table)
```

```python
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 4))
plt.title(f'Asset allocation vs Asset currency (frequency)')
plt.ylabel('Frequency on "asset allocation"')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
contingency_table_norm = contingency_table.div(contingency_table.sum(axis=1), axis=0)
contingency_table_norm.plot(kind='bar', stacked=True, figsize=(10, 4))
plt.title(f'Asset allocation vs Asset currency (proportions)')
plt.ylabel('Proportion on "asset allocation"')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
contingency_table = pd.crosstab(df['asset_currency'], df['asset_allocation'])
contingency_table_norm = contingency_table.div(contingency_table.sum(axis=1), axis=0)
contingency_table_norm.plot(kind='bar', stacked=True, figsize=(10, 4))
plt.title(f'Asset allocation vs Asset currency (proportions)')
plt.ylabel('Proportion on "asset currency"')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_corr = df[num_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(df_corr, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numerical Features')
plt.tight_layout()
plt.show()
```
